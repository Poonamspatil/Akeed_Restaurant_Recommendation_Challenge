## Import libraries

```python
import pandas as pd
```

## Set filenames

```python
orders=pd.read_csv("Data/orders.csv")
train_locations=pd.read_csv("Data/train_locations.csv")
train_customers=pd.read_csv("Data/train_customers.csv")
vendors=pd.read_csv("Data/vendors.csv")
SS=pd.read_csv("Data/SampleSubmission.csv")
```

## Remove customer duplicate info

```python
train_customers.head() #Customer id details
```

```python
len(train_customers)
```

```python
train_customers['akeed_customer_id'].value_counts() #There are duplicate values, let's drop them
```

```python
train_customers.drop_duplicates(subset='akeed_customer_id',keep='first',inplace=True)
```

```python
print(train_customers['akeed_customer_id'].nunique())
print(len(train_customers))
```

## Remove customer locations duplicate info

```python
train_locations.head() #details on customer location
```

```python
print(len(train_locations))
print(train_locations['customer_id'].value_counts())
```

```python
train_locations.drop_duplicates(subset='customer_id',keep='first',inplace=True)
```

```python
print(train_locations['customer_id'].nunique())
print(len(train_locations))
```

## Merge customer and locations

```python
cust_loc=pd.merge(train_customers,train_locations,how="left",left_on='akeed_customer_id',right_on='customer_id')
del cust_loc['customer_id']
```

```python
cust_loc
```

## Order dataset

```python
orders #orders made by the customers in train_customer dataset
```

## Create target variable based on order status

```python
orders['Target']=1
```

## Merge customer details and order details

```python
cust_loc_order=pd.merge(cust_loc,orders,how="left",left_on='akeed_customer_id',right_on='customer_id')
```

```python
cust_loc_order
```

```python
cust_loc_order['Target']=cust_loc_order['Target'].fillna(0)
```

```python
cust_loc_order
```

## Merge customer order and vendor details

```python
vendors #vendors that customers can order from
```

```python
print(vendors['id'].value_counts())
```

```python
cust_loc_order_vendor=pd.merge(cust_loc_order,vendors,how="left",left_on='vendor_id',right_on='id')
```

```python
cust_loc_order_vendor.head()
```

```python
cust_loc_order_vendor.columns.values
```

```python
cust_loc_order_vendor.shape
```

```python
cust_loc_order_vendor.to_csv("train.csv",index=False)
```
